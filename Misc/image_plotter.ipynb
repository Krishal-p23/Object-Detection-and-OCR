```python
import cv2
import json
from matplotlib import pyplot as plt
```

```python
img_path = 'dataset/testing_data/images/87093315_87093318.png'
label_path = 'dataset/testing_data/annotations/87093315_87093318.json'
```

```python
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
img.shape
```

```python
with open(label_path, 'r') as f:
    anns = json.load(f)
    
for ann in anns["form"]:
    words = ann["words"]
    for word in words:
        bbox = word["box"]
        text = word["text"]
        img = cv2.rectangle(img, bbox[:2], bbox[2:], (255, 0, 0), 2)
        img = cv2.putText(img, text, bbox[:2], cv2.FONT_HERSHEY_PLAIN, 0.8, (0, 255, 0), 1)
        
plt.imshow(img)
cv2.imwrite("ann.png", img)
```
