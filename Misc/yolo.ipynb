```python
from ultralytics import YOLO
```

```python
yolo_model = YOLO('YOLO/yolo_training/train2/weights/best.pt')
```

```python
img = 'C:/Users/hp/OneDrive/Desktop/NIC/YOLO/people.jpg'
```

```python
results = yolo_model.predict(img, conf=0.25)
object_names = []
if results and results[0].boxes is not None:
        for box in results[0].boxes.data.tolist():
            cls_id = int(box[5])
            object_names.append(results[0].names[cls_id])
            
set(object_names)
```

```python
import text_ocr
```

```python
if 'text' in object_names:
    text = text_ocr.extract_text_from_image(img)
    print(text)
else:
    print(" ".join(object_names))
```
